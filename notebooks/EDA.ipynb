```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart_cleveland_upload.csv")
```

```python
target = df["condition"]
```

```python
correlations = df.corr()

sns.set(rc = {'figure.figsize':(16,8)})
sns.heatmap(correlations).set_title("Correlations")
```

```python
sns.set(rc = {'figure.figsize':(16,8)})
sns.pairplot(df, hue = "condition")
```

```python
df["condition"].hist()
plt.title("Distribution of target")
plt.show()
```

```python
correlations_phik = df.phik_matrix()
sns.set(rc = {'figure.figsize':(16,8)})
sns.heatmap(correlations_phik).set_title("Phi_k correlation")
```
